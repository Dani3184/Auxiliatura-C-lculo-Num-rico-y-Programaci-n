**Lectura y escritura de archivos**

```python
# Crear datos
numeros = [1, 2, 3, 4, 5]

# Escribir archivo
with open("datos.txt", "w") as archivo:
    for n in numeros:
        archivo.write(str(n) + "\n")
```

```python
numeros = []

with open("datos.txt", "r") as archivo:
    for linea in archivo:
        print(linea)
        numeros.append(float(linea.strip()))
print(numeros)
```

```python
import math

with open("tabla.txt", "w") as archivo:
    for i in range(10):
        x = i * 0.1
        y = math.sin(x)
        archivo.write(f"{x} {y}\n")
```

```python
x_datos = []
y_datos = []

with open("tabla.txt", "r") as archivo:
    for linea in archivo:
        valores = linea.split()
        print(valores)
        x_datos.append(float(valores[0]))
        y_datos.append(float(valores[1]))

print(x_datos)
print(y_datos)
```

**Graficos simples**


```python
import matplotlib.pyplot as plt

plt.plot(x_datos, y_datos)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Grafica desde archivo")
plt.show()
```
