# Gráficas

Para crear un gráfico con `matplotlib` es habitual seguir los siguientes pasos:

1. **Importar el módulo `pyplot`.**  
   Este módulo contiene las funciones necesarias para generar gráficos.

2. **Crear la figura y los ejes.**  
   La figura es la ventana o región donde se dibuja el gráfico, y los ejes son el espacio donde se representan los datos.  
   Para ello se utiliza la función `subplots()`.

3. **Dibujar los datos sobre los ejes.**  
   Dependiendo del tipo de gráfico que se desee (línea, puntos, barras, etc.), se emplean distintas funciones.

4. **Personalizar el gráfico.**  
   Se pueden añadir título, etiquetas de los ejes, leyenda, rejilla, colores y otras configuraciones para mejorar la presentación.

5. **Guardar el gráfico (opcional).**  
   Se utiliza la función `savefig()` para almacenar la imagen en un archivo.

6. **Mostrar el gráfico.**  
   Finalmente, se emplea la función `show()` para visualizarlo en pantalla.

**Forma típica**

```python
import matplotlib.pyplot as plt
import math

x = []
y = []

for i in range(-20, 21):
    valor = i * 0.1
    x.append(valor)
    y.append(math.exp(-valor**2))

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Funcion e^{-x^2}")
plt.grid()
plt.savefig("hola.jpg")
plt.show()
```

# Práctica: Proyectil con Resistencia del Aire

## 1. Fundamentos Matemáticos
Cuando se considera el efecto del aire, las coordenadas del proyectil en función del tiempo $t$ están dadas por:

$$x(t) = \frac{v_{0x}}{b}(1 - e^{-bt})$$
$$y(t) = \frac{1}{b^2}(g + bv_{0y})(1 - e^{-bt}) - \frac{g}{b}t$$

Donde:
* $b$: Coeficiente de arrastre.
* $g$: Aceleración de la gravedad ($9.8 \, m/s^2$).
* $v_{0x}, v_{0y}$: Componentes de la velocidad inicial.

Las componentes de la **velocidad** y la **aceleración** son:
* $v_x = v_{0x}e^{-bt}$
* $v_y = \frac{1}{b}(g + bv_{0y})e^{-bt} - \frac{g}{b}$
* $a_x = -bv_{0x}e^{-bt}$
* $a_y = -(g + bv_{0y})e^{-bt}$

---

## 2. Objetivo de la Simulación
Escriba un programa para investigar el vuelo de un proyectil con:
* **Velocidad inicial ($v_0$):** $50 \, m/s$
* **Ángulo de elevación:** $25^\circ$
* **Intervalo de tiempo:** de $t = 0$ a $4.5 \, s$ con pasos de $0.1 \, s$.

### Tareas a realizar:
1. **Punto más elevado:** Hallar coordenadas, velocidad y aceleración en el punto máximo para $b = 0.10 \, s^{-1}$.
2. **Influencia del aire:** Comparar el alcance (valor de $x$ cuando $y=0$) y la velocidad final usando dos coeficientes:
    * Case A: $b = 0.10 \, s^{-1}$
    * Case B: $b = 0.20 \, s^{-1}$
3. **Análisis:** Explicar por qué en el punto más elevado $a_y = -g$ y por qué $a_x$ nunca es cero.

```python
#Codigo para calcular
# Solucion 
import math as mt
import matplotlib.pyplot as plt
g = 9.81
b1 = 0.1
b2 = 0.2
alfa = 45
v0 = 50
def posx(t,b):
    alfarad = alfa*mt.pi/180
    v0x = v0*mt.cos(alfarad)
    x = v0x*(1-mt.exp(-b*t))/b
    return x
def posy(t,b):
    alfarad = alfa*mt.pi/180
    v0y = v0*mt.sin(alfarad)
    y = (1-mt.exp(-b*t))*(g+b*v0y)/b**2 - g*t/b
    return y

def velx(t,b):
    alfarad = alfa*mt.pi/180
    v0x = v0*mt.cos(alfarad)
    vx = v0x*mt.exp(-b*t) 
    return vx   
def vely(t,b):
    alfarad = alfa*mt.pi/180
    v0y = v0*mt.sin(alfarad)
    vy = (g + b*v0y)*mt.exp(-b*t)/b - g/b
    return vy  
def alx(t,b):
    alfarad = alfa*mt.pi/180
    v0x = v0*mt.cos(alfarad)
    ax = -b*v0x*mt.exp(-b*t)
    return ax
def aly(t,b):
    alfarad = alfa*mt.pi/180
    v0y = v0*mt.sin(alfarad)
    ay = -(g+b*v0y)*mt.exp(-b*t)
    return ay
X1 = []
Y1 = []
VX1 = []
VY1 = []
AX1 = []
AY1 = []
T = []

X2 = []
Y2 = []
VX2 = []
VY2 = []
AX2 = []
AY2= []
t = 0
while t<=4.5:
    T.append(t)
    X1.append(posx(t,b1))
    Y1.append(posy(t,b1))
    VX1.append(velx(t,b1))
    VY1.append(vely(t,b1))
    AX1.append(alx(t,b1))
    AY1.append(aly(t,b1))
    X2.append(posx(t,b2))
    Y2.append(posy(t,b2))
    VX2.append(velx(t,b2))
    VY2.append(vely(t,b2))
    AX2.append(alx(t,b2))
    AY2.append(aly(t,b2))
    t+=0.1

fig, axs = plt.subplots(3, 2, figsize=(12, 10), sharex=True)
fig.suptitle('Análisis Cinemático de un Proyectil', fontsize=16)

# --- Gráficos de Posición ---
axs[0, 0].plot(T, X1, color='blue', label = 'b = 0.1')
axs[0, 0].plot(T, X2, color='red', label = 'b = 0.2')
axs[0, 0].set_ylabel('Posición X [m]')
axs[0, 0].grid(True)

axs[0, 1].plot(T, Y1, color='blue',label = 'b = 0.1')
axs[0, 1].plot(T, Y2, color='red',label = 'b = 0.2')
axs[0, 1].set_ylabel('Posición Y [m]')
axs[0, 1].grid(True)

# --- Gráficos de Velocidad ---
axs[1, 0].plot(T, VX1, color='blue', linestyle='--',label = 'b = 0.1')
axs[1, 0].plot(T, VX2, color='red', linestyle='--',label = 'b = 0.2')
axs[1, 0].set_ylabel('Velocidad VX [m/s]')
axs[1, 0].grid(True)

axs[1, 1].plot(T, VY1, color='blue', linestyle='--',label = 'b = 0.1')
axs[1, 1].plot(T, VY2, color='red', linestyle='--',label = 'b = 0.2')
axs[1, 1].set_ylabel('Velocidad VY [m/s]')
axs[1, 1].grid(True)

# --- Gráficos de Aceleración ---
axs[2, 0].plot(T, AX1, color='blue', drawstyle='steps-post',label = 'b = 0.1')
axs[2, 0].plot(T, AX2, color='red', drawstyle='steps-post',label = 'b = 0.2')
axs[2, 0].set_ylabel('Aceleración AX [m/s²]')
axs[2, 0].set_xlabel('Tiempo [s]')
axs[2, 0].grid(True)

axs[2, 1].plot(T, AY1, color='blue', drawstyle='steps-post',label = 'b = 0.1')
axs[2, 1].plot(T, AY2, color='red', drawstyle='steps-post',label = 'b = 0.2')
axs[2, 1].set_ylabel('Aceleración AY [m/s²]')
axs[2, 1].set_xlabel('Tiempo [s]')
axs[2, 1].grid(True)

for ax in axs.flat:
    ax.legend(fontsize='small', loc='upper right')
# Ajustar el diseño para que no se traslapen las etiquetas
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

plt.plot(X1,Y1, color = 'blue', label="b = 0.1")
plt.plot(X2,Y2, color = 'red', label="b = 0.2")
plt.legend()
plt.show()
```

```python
import math as mt
import pandas as pd
import matplotlib.pyplot as plt

# Constantes
g = 9.81
b1 = 0.1
b2 = 0.2
alfa = 45
v0 = 50

alfarad = alfa * mt.pi / 180
v0x = v0 * mt.cos(alfarad)
v0y = v0 * mt.sin(alfarad)

# Crear lista de tiempos
t = [i * 0.1 for i in range(46)]

# Crear DataFrame
df = pd.DataFrame({"t": t})

# ----- POSICION -----
df["x1"] = v0x * (1 - mt.e**(-b1*df["t"])) / b1
df["x2"] = v0x * (1 - mt.e**(-b2*df["t"])) / b2

df["y1"] = (1 - mt.e**(-b1*df["t"])) * (g + b1*v0y) / b1**2 - g*df["t"]/b1
df["y2"] = (1 - mt.e**(-b2*df["t"])) * (g + b2*v0y) / b2**2 - g*df["t"]/b2

# ----- VELOCIDAD -----
df["vx1"] = v0x * mt.e**(-b1*df["t"])
df["vx2"] = v0x * mt.e**(-b2*df["t"])

df["vy1"] = (g + b1*v0y) * mt.e**(-b1*df["t"]) / b1 - g/b1
df["vy2"] = (g + b2*v0y) * mt.e**(-b2*df["t"]) / b2 - g/b2

# ----- ACELERACION -----
df["ax1"] = -b1 * v0x * mt.e**(-b1*df["t"])
df["ax2"] = -b2 * v0x * mt.e**(-b2*df["t"])

df["ay1"] = -(g + b1*v0y) * mt.e**(-b1*df["t"])
df["ay2"] = -(g + b2*v0y) * mt.e**(-b2*df["t"])

# ----- GRAFICOS -----

fig, axs = plt.subplots(3, 2, figsize=(12, 10), sharex=True)
fig.suptitle('Análisis Cinemático de un Proyectil', fontsize=16)

# Posición X
axs[0,0].plot(df["t"], df["x1"], label="b = 0.1")
axs[0,0].plot(df["t"], df["x2"], label="b = 0.2")
axs[0,0].set_ylabel("Posición X [m]")
axs[0,0].grid()

# Posición Y
axs[0,1].plot(df["t"], df["y1"], label="b = 0.1")
axs[0,1].plot(df["t"], df["y2"], label="b = 0.2")
axs[0,1].set_ylabel("Posición Y [m]")
axs[0,1].grid()

# Velocidad X
axs[1,0].plot(df["t"], df["vx1"], "--", label="b = 0.1")
axs[1,0].plot(df["t"], df["vx2"], "--", label="b = 0.2")
axs[1,0].set_ylabel("Velocidad VX [m/s]")
axs[1,0].grid()

# Velocidad Y
axs[1,1].plot(df["t"], df["vy1"], "--", label="b = 0.1")
axs[1,1].plot(df["t"], df["vy2"], "--", label="b = 0.2")
axs[1,1].set_ylabel("Velocidad VY [m/s]")
axs[1,1].grid()

# Aceleración X
axs[2,0].plot(df["t"], df["ax1"], label="b = 0.1")
axs[2,0].plot(df["t"], df["ax2"], label="b = 0.2")
axs[2,0].set_ylabel("Aceleración AX [m/s²]")
axs[2,0].set_xlabel("Tiempo [s]")
axs[2,0].grid()

# Aceleración Y
axs[2,1].plot(df["t"], df["ay1"], label="b = 0.1")
axs[2,1].plot(df["t"], df["ay2"], label="b = 0.2")
axs[2,1].set_ylabel("Aceleración AY [m/s²]")
axs[2,1].set_xlabel("Tiempo [s]")
axs[2,1].grid()

for ax in axs.flat:
    ax.legend()

plt.tight_layout()
plt.show()

# ----- TRAYECTORIA -----

plt.plot(df["x1"], df["y1"], label="b = 0.1")
plt.plot(df["x2"], df["y2"], label="b = 0.2")

plt.xlabel("X [m]")
plt.ylabel("Y [m]")
plt.title("Trayectoria del proyectil")
plt.legend()
plt.grid()
plt.show()

print(df)
```
