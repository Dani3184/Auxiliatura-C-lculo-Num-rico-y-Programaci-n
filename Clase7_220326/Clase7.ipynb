**🌻 Proyecto 1: Modelado de Filotaxis (Girasol de Fermat)**

**Algoritmo**

Siga estas instrucciones para construir su programa en Python utilizando las librerías `numpy` y `matplotlib`:

1.  **Definición de Constantes:**
    * Cree una variable `N` con el valor de `1000` (número total de semillas).
    * Calcule el **Número de Oro** ($\phi$) usando la fórmula: $\phi = \frac{1 + \sqrt{5}}{2}$.
    * Calcule el **Ángulo de Oro** en radianes usando la relación: $\alpha = 2\pi \cdot (1 - \frac{1}{\phi})$.

2.  **Creación de Arreglos (Vectores):**
    * **Índices:** Cree un arreglo de NumPy llamado `indices` que contenga números enteros desde `0` hasta $N-1$.
    * **Ángulos:** Cree un arreglo `angulos` que sea el resultado de multiplicar cada índice por el **Ángulo de Oro** ($\theta_i = i \cdot \alpha$).
    * **Radios:** Cree un arreglo `radios` calculando la raíz cuadrada de cada índice ($r_i = \sqrt{i}$).

3.  **Conversión de Coordenadas:**
    * Transforme sus datos de coordenadas polares $(r, \theta)$ a coordenadas cartesianas $(x, y)$ mediante las funciones:
        * $x = r \cdot \cos(\theta)$
        * $y = r \cdot \sin(\theta)$

4.  **Visualización de Datos:**
    * Utilice `plt.scatter(x, y)` para graficar.
    * **Configuración:** Use `c=indices` para el color, `cmap='viridis'` para la paleta y `plt.axis('equal')` para mantener la proporción circular.



```python
N = 1000
import math as mt
oro = (1+mt.sqrt(5))/2
ang_oro = 2*mt.pi*(1-1/oro)
import matplotlib.pyplot as plt
import numpy as np
vec = np.arange(0, N)

ang_puntos = ang_oro*vec

radios = np.sqrt(vec)

# print(vec)
# print(radios)
# print(ang_puntos)

##Coordenadas polares
x = radios*np.cos(ang_puntos)
y = radios*np.sin(ang_puntos)

plt.figure(figsize=(9,9), facecolor='black' )

plt.scatter(x,y, c=vec,cmap='jet', s=vec/5 +10, alpha = 0.9)
plt.savefig("HOLA.png")
plt.show
```

```python
#media por bloques

ventana = 5
vec = df['DS18B20_0'].to_numpy()
print(vec)
n = len(vec)
mediabloqu = []
for i in range(0,n,ventana):
    xdd = vec[i:i+ventana]
    if len(xdd) == ventana:
        prom = float(np.sum(xdd)/ventana)
        mediabloqu.append(prom)
print(n)
print(mediabloqu)  
print(len(mediabloqu))  
```
