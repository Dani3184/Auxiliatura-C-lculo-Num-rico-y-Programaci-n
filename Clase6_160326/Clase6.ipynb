*Numpy*

Numpy es una librería de Python para computación científica, Numpy significa NumericalPython.

Según su página web, es fundamental para computación científica con python, y contiene:

    * Un poderoso objeto de arreglo N dimensional

    * Funciones (radiodifusión) sofisticadas

    * Herramientas para integrar código en C/C++ y Frotran
    
    * Útiles capacidades de álgebra lineal, transformación de Fourier y números aleatorios

Incluso Numpy puede ser utilizado como un contenedor multidimensional de datos. 

La novedad de Numpy es la incorporación de los **arrays**, facilitando el uso de las listas (vectores) habituales de python. 

```python
#Los nuevos arreglos (array de Numpy)
import numpy as np
vector = [1, 2, 3, 4, 5]

arregloNp = np.array(vector)
print(vector)
print(arregloNp) #note que se imprime diferente
```

**Ejercicio 1 — Distribución Normal**

En muchos experimentos físicos, los errores de medición siguen una **distribución normal**.

1. Generar 1000 números aleatorios usando una distribución normal con media $\mu = 0$ y desviación estándar $\sigma = 1$.
2. Graficar un **histograma** de los datos.
3. Calcular la **media** y la **desviación estándar** de la muestra.
4. Superponer sobre el histograma la **función de densidad de la distribución normal**.

La función de densidad de la distribución normal está dada por:

$$
f(x)=\frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}
$$
