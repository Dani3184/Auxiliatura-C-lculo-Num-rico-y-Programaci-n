 Operaciones con pandas
 
Se pueden realizar diversas operaciones con los dataFrames y Series que nos ofrece pandas.
Nos permite operar sobre las columnas sin utilizar bucles:

```python
import pandas as pd
df = pd.DataFrame({
    'tiempo': [0, 1, 2, 3, 4, 5],
    'distancia': [0, 3, 6, 9, 12, 15]
})
print(df)
df["velocidad"] = df["distancia"] / df["tiempo"]
print(df)

#otra forma
df["Velocidad2"] = df["distancia"].diff()
# print(df)
df["Distancias cuadradas"] = df["distancia"]**2
print(df)
# df["distanciasXD"] = df["distancia"] > 20
# print(df)
filtro = df["distancia"]>2
print(pd.Series(filtro))

suma = df["distancia"].sum()
print(suma)
```

```python
suma = df["distancia"].sum()
media = df["distancia"].mean()
mediana = df["distancia"].median()
max = df["distancia"].max()

print(suma, media, mediana, max, sep="--")
#Grafico con matplotlib
import matplotlib.pyplot as plt
plt.plot(df["tiempo"], df["distancia"])
plt.grid()
plt.show()
#Con pandas
df.plot(x = "tiempo", y = "distancia")
```

```python
#Utilizando pandas tenemos:
posiciones = {
    "x":[13, 12, 11, 18, 17, 14,16,18,11,13],
    "y":[75,72,74,90,94,76,90,98,65,88]
}

df = pd.DataFrame(posiciones)

sumax = df["x"].sum()
sumay = df["y"].sum()
sumaxsquare = (df["x"]**2).sum()
sumaxy = (df["x"]*df["y"]).sum()
n = df["x"].size
mediay = df["y"].mean()
mediax = df["x"].mean()
Sxx= sumaxsquare - (sumax**2)/n
Sxy = sumaxy - (sumax*sumay)/n

b = Sxy/Sxx
a = mediay - b*mediax

print(a,b, sep="---")
#Calculamos con los valores de la regresion
yreg = a+b*df["x"]
#Grafico


plt.figure(figsize=(5,5))

plt.scatter(df["x"],df["y"], colorizer="blue",label="Puntos")
plt.plot(df["x"],yreg, color = "red", label="regresion")
plt.legend()
plt.grid()
plt.show()
```
