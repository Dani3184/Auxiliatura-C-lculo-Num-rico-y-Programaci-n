Práctica de Análisis: Espectros de Energía y Flujo Diferencial

1. Contexto y Objetivos
El objetivo de esta práctica es procesar datos crudos de una simulación de lluvias atmosféricas extensas para caracterizar el comportamiento energético de las partículas secundarias (gammas, electrones, positrones, muones, neutrones y protones). 

Se busca transformar la información de **Energía Cinética ($E_k$)** en **Energía Total ($E_{tot}$)** y calcular el **Flujo Diferencial ($dN/dE$)** para cada especie.

2. Definiciones Teóricas
Para cada partícula detectada con un identificador `id`, se deben aplicar las siguientes consideraciones:

* **Energía Total:** Se calcula sumando la energía cinética y la masa en reposo de la partícula:
    $$E_{tot} = E_k + m_0 c^2$$
* **Masas de Referencia ($m_0 c^2$):**
    * **Fotones (ID 1):** $0$
    * **Electrones/Positrones (ID 2, 3):** $0.511 \text{ MeV}$
    * **Muones (ID 5, 6):** $105.66 \text{ MeV}$
    * **Neutrones (ID 13):** $939.57 \text{ MeV}$
    * **Protones (ID 14):** $938.27 \text{ MeV}$

3. Metodología de Análisis
El algoritmo de procesamiento debe seguir estos pasos de manera vectorizada para optimizar el rendimiento:

1.  **Filtrado por Especie:** Separar los eventos según su `id`.
2.  **Cálculo de Bines Adaptativos:** * Identificar $E_{min}$ y $E_{max}$ para cada tipo de partícula.
    * Definir un ancho de bin constante: $\Delta E = \frac{E_{max} - E_{min}}{100}$.
3.  **Normalización del Histograma (Flujo):** * Contar el número de eventos en cada intervalo.
    * Dividir el conteo de cada bin entre su ancho ($\Delta E$) para obtener el flujo diferencial:
    $$\frac{dN}{dE} \approx \frac{\text{counts}}{\Delta E}$$
4.  **Visualización:** Generar gráficos de flujo vs. energía total, utilizando escalas logarítmicas cuando el rango dinámico de los datos lo requiera.
